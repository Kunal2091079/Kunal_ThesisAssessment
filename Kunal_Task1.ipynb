```python
#Task 1A

from bokeh.plotting import figure, show, output_notebook
from bokeh.models import ColumnDataSource
import pandas as pd

fill_colors = []
border_colors = []

for index, row in df.iterrows():
    if row["label"] == "dog":
        if row["classifierA_predicted_label"] == "dog" and row["classifierB_predicted_label"] == "dog":
            fill_colors.append("black")
        elif row["classifierA_predicted_label"] == "dog" or row["classifierB_predicted_label"] == "dog":
            fill_colors.append("gray")
        else:
            fill_colors.append("white")
        border_colors.append("red")
    else:
        if row["classifierA_predicted_label"] == "cat" and row["classifierB_predicted_label"] == "cat":
            fill_colors.append("black")
        elif row["classifierA_predicted_label"] == "cat" or row["classifierB_predicted_label"] == "cat":
            fill_colors.append("gray")
        else:
            fill_colors.append("white")
        border_colors.append("blue")

df["fill_colors"] = fill_colors
df["border_colors"] = border_colors

source = ColumnDataSource(df)

p = figure(width=800, height=600)

p.circle(
    x="x",
    y="y",
    source=source,
    size=8,
    fill_color="fill_colors",
    line_color="border_colors",
    line_width=2,
)

p.title.text = "Classifier A vs Classifier B Accuracy Scatter Plot"
p.xaxis.axis_label = "X Coordinate"
p.yaxis.axis_label = "Y Coordinate"

output_notebook()
show(p)
```

```python
from bokeh.plotting import figure, output_file, show
import pandas as pd

df = pd.read_csv('Synthetic_2_classifiers.csv')

def second():
    df['Allcorrect'] = (df['label'] == df['classifierA_predicted_label']) & (df['label'] == df['classifierB_predicted_label'])

    Total_Red = (df['label'] == 'dog').sum()
    Total_Blue = (df['label'] == 'cat').sum()

    correct_predictions_A = df['classifierA_predicted_label'] == df['label']
    correct_predictions_B = df['classifierB_predicted_label'] == df['label']

    count_dog_A = ((correct_predictions_A) & (df['label'] == 'dog')).sum()
    count_dog_B = ((correct_predictions_B) & (df['label'] == 'dog')).sum()

    count_cat_A = ((correct_predictions_A) & (df['label'] == 'cat')).sum()
    count_cat_B = ((correct_predictions_B) & (df['label'] == 'cat')).sum()

    A = ((df['Allcorrect']) & (df['label'] == 'dog')).sum()
    B = ((df['Allcorrect']) & (df['label'] == 'dog')).sum()
    C = ((df['Allcorrect']) & (df['label'] == 'cat')).sum()
    D = ((df['Allcorrect']) & (df['label'] == 'cat')).sum()

    categories = ['Classifier A Correctly predicted dogs', 'Classifier B Correctly predicted dogs',
                  'Classifier A Correctly predicted cats', 'Classifier B Correctly predicted cats']

    p = figure(x_range=categories, title="Bar Chart Example", width=900)

    bar_width = 0.2

    p.vbar(x=[0.7, 1.3, 2.7, 3.3], top=[A, B, C, D], width=bar_width, fill_color="gray", line_color="gray", legend_label="Total Predictions")

    p.vbar(x=[1, 1], top=[Total_Red, Total_Red], width=1, fill_color="lightcoral", line_color="", line_width=1, legend_label="Total Red", alpha=0.10)
    p.vbar(x=[3, 3], top=[Total_Blue, Total_Blue], width=1, fill_color="lightblue", line_color="", line_width=1, legend_label="Total Blue", alpha=0.10)

    p.vbar(x=[0.7, 1.3, 2.7, 3.3], top=[count_dog_A, count_dog_B, count_cat_A, count_cat_B], width=bar_width, fill_color="black", line_color="black", alpha=0.7, legend_label="Correct Predictions")

    p.xaxis.axis_label = "Categories"
    p.yaxis.axis_label = "Count"
    p.title.align = "center"
    p.title.text_font_size = "20px"
    p.legend.title = "Legend"

    show(p)

second()
```

```python
#Task 1C

import pandas as pd
from bokeh.plotting import figure, show, ColumnDataSource
from bokeh.layouts import column
from bokeh.models import LassoSelectTool, HoverTool, TapTool, CustomJS, ColorBar
from bokeh.transform import factor_cmap
from bokeh.io import output_notebook

data = pd.read_csv("Synthetic_2_classifiers.csv")

dogs = data[data['label'] == 'dog']
cats = data[data['label'] == 'cat']

dog_both_classifiers = len(dogs[(dogs['label'] == 'dog') &
                                (dogs['classifierA_predicted_label'] == 'dog') &
                                (dogs['classifierB_predicted_label'] == 'dog')])

dog_one_classifier = len(dogs[((dogs['label'] == 'dog') &
                               ((dogs['classifierA_predicted_label'] == 'dog') |
                                (dogs['classifierB_predicted_label'] == 'dog'))) &
                              ~((dogs['classifierA_predicted_label'] == 'dog') &
                                (dogs['classifierB_predicted_label'] == 'dog'))])

cat_both_classifiers = len(cats[(cats['label'] == 'cat') &
                                (cats['classifierA_predicted_label'] == 'cat') &
                                (cats['classifierB_predicted_label'] == 'cat')])

cat_one_classifier = len(cats[((cats['label'] == 'cat') &
                               ((cats['classifierA_predicted_label'] == 'cat') |
                                (cats['classifierB_predicted_label'] == 'cat'))) &
                              ~((cats['classifierA_predicted_label'] == 'cat') &
                                (cats['classifierB_predicted_label'] == 'cat'))])

output_notebook()

scatter_source_dog = ColumnDataSource(data=dogs)
scatter_source_cat = ColumnDataSource(data=cats)

color_map = factor_cmap(field_name='label', palette=['red', 'blue'], factors=['dog', 'cat'])

scatter_plot = figure(plot_width=600, plot_height=400, title="Scatter Plot",
                      tools=[LassoSelectTool(), HoverTool(tooltips=[("Label", "@label"),
                                                                    ("Classifier A", "@classifierA_predicted_label"),
                                                                    ("Classifier B", "@classifierB_predicted_label")]),
                             TapTool()])

scatter_plot.scatter('x', 'y', source=scatter_source_dog, size=10, color=color_map, legend_label='Dogs')
scatter_plot.scatter('x', 'y', source=scatter_source_cat, size=10, color=color_map, legend_label='Cats', marker='square')

bar_source = ColumnDataSource(data={'classifiers': ['Classifier A', 'Classifier B'],
                                    'both': [0, 0],
                                    'one': [0, 0],
                                    'none': [0, 0]})
bar_plot = figure(x_range=['Classifier A', 'Classifier B'], plot_height=400, plot_width=600, title="Classifier Performance",
                  toolbar_location=None, tools="")

bar_plot.vbar(x='classifiers', top='both', source=bar_source, width=0.3, color='black', legend_label='Both Classifiers')
bar_plot.vbar(x='classifiers', top='one', source=bar_source, width=0.3, color='gray', legend_label='One Classifier',
              bottom='both')
bar_plot.vbar(x='classifiers', top='none', source=bar_source, width=0.3, color='gray', legend_label='None Classified',
              bottom='both')

bar_plot.x_range.range_padding = 0.1
bar_plot.xgrid.grid_line_color = None
bar_plot.legend.title = 'Categories'
bar_plot.legend.label_text_font_size = '12pt'

callback = CustomJS(args=dict(scatter_source_dog=scatter_source_dog, scatter_source_cat=scatter_source_cat, bar_source=bar_source), code="""
    var selected_indices_dog = scatter_source_dog.selected.indices;
    var selected_indices_cat = scatter_source_cat.selected.indices;
    
    var both_count_A = 0;
    var both_count_B = 0;
    var one_count_A = 0;
    var one_count_B = 0;
    var none_count_A = 0;
    var none_count_B = 0;
    
    function countClassifier(classifier, label) {
        var count = 0;
        for (var i = 0; i < classifier.length; i++) {
            var index = classifier[i];
            var predictedLabel = label[index];
            if (predictedLabel === 'dog') {
                count += 1;
            }
        }
        return count;
    }
    
    both_count_A = countClassifier(selected_indices_dog, scatter_source_dog.data['classifierA_predicted_label']);
    both_count_B = countClassifier(selected_indices_dog, scatter_source_dog.data['classifierB_predicted_label']);
    both_count_A += countClassifier(selected_indices_cat, scatter_source_cat.data['classifierA_predicted_label']);
    both_count_B += countClassifier(selected_indices_cat, scatter_source_cat.data['classifierB_predicted_label']);
    
    one_count_A = countClassifier(selected_indices_dog, scatter_source_dog.data['classifierA_predicted_label']);
    one_count_B = countClassifier(selected_indices_dog, scatter_source_dog.data['classifierB_predicted_label']);
    one_count_A += countClassifier(selected_indices_cat, scatter_source_cat.data['classifierA_predicted_label']);
    one_count_B += countClassifier(selected_indices_cat, scatter_source_cat.data['classifierB_predicted_label']);
    
    none_count_A = selected_indices_dog.length + selected_indices_cat.length - both_count_A - one_count_A;
    none_count_B = selected_indices_dog.length + selected_indices_cat.length - both_count_B - one_count_B;
    
    bar_source.data['both'] = [both_count_A, both_count_B];
    bar_source.data['one'] = [one_count_A, one_count_B];
    bar_source.data['none'] = [none_count_A, none_count_B];
    bar_source.change.emit();
""")

scatter_source_dog.selected.js_on_change('indices', callback)
scatter_source_cat.selected.js_on_change('indices', callback)

layout = column(scatter_plot, bar_plot)

show(layout)
```
